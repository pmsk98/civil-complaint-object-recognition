```python
labels_dict = {
    "1":0, "2":0, "3":1, "4":1,
    "5":0, "6":0, "7":1, "8":1,
    "9":2, "10":2, "11":3, "12":3,
    "19":4, "20":4, "21":5, "22":5,
    "23":6, "24":6, "25":7, "26":7,
    "27":8, "28":8, "29":9, "30":9
}
```

```python
def json2txt(json_path, txt_path):
    with open(json_path, "r", encoding="UTF-8") as f:
        json_data = json.load(f)
        
    boxs = json_data["annotations"]["Bbox Annotation"]["Box"]
    
    box_list = []
    for box in boxs:
        row = str(labels_dict[str(box["category_id"])])+" "+str(box["x"])+" "+str(box["y"])+" "+str(box["w"])+" "+str(box["h"])
        box_list.append(row)
        
    with open(txt_path, "w") as f:
        f.write("\n".join(box_list)+"\n")
```

```python
file_path = "../darknet/build/darknet/x64/data/obj/"
```

```python
json_list = glob(file_path+"*.json")
```

```python
json_list
```

```python
one_split = list(map(lambda x: x.split("\\")[1], json_list))
one_split
```

```python
data_names = list(map(lambda x: x.split(".")[0], one_split))
data_names
```

```python
for d in data_names: json2txt(file_path+d+".json", file_path+d+".txt")
```
